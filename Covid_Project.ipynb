```python
# Preparing data for analysis
```

```python
df1.columns
```

```python
df_copy=df1
```

```python
X=df1[['Breathing Problem', 'Fever', 'Dry Cough', 'Sore throat',
       'Running Nose', 'Asthma', 'Chronic Lung Disease', 'Headache',
       'Heart Disease', 'Diabetes', 'Hyper Tension', 'Fatigue ',
       'Gastrointestinal ', 'Abroad travel', 'Contact with COVID Patient',
       'Attended Large Gathering', 'Visited Public Exposed Places',
       'Family working in Public Exposed Places']]
y=df1['COVID-19']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.6 ,test_size = 0.4, random_state=100)
```

```python
# LogisticRegression()
```

```python
from sklearn.linear_model import LogisticRegression
logmodel1 = LogisticRegression()
logmodel1.fit(X_train,y_train)
```

```python
predictions1 = logmodel1.predict(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
result=confusion_matrix(y_test,predictions1)
print(result)
```

```python
print(classification_report(y_test,predictions1))
```

```python
print(' Accuracy: ',accuracy_score(y_test, predictions1)*100,'\n')
```

```python
#Support Vector Classification model
```

```python
from sklearn.svm import SVC
svc_model2 = SVC()
svc_model2.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_predict2 = svc_model2.predict(X_test)
cm2 = confusion_matrix(y_test, y_predict2)
print(classification_report(y_test, y_predict2))
```

```python
print(' Accuracy: ',accuracy_score(y_test, y_predict2)*100,'\n')
```

```python
#random forest
```

```python
from sklearn.ensemble import RandomForestClassifier
model3 = RandomForestClassifier(n_estimators=40)
model3.fit(X_train, y_train)
```

```python
model3.score(X_test,y_test)
```

```python
### Manual Hyperparameter Tuning
```

```python
model4=RandomForestClassifier(n_estimators=300,criterion='entropy',
                             max_features='sqrt',min_samples_leaf=10,random_state=100).fit(X_train,y_train)
predictions4=model4.predict(X_test)
print(confusion_matrix(y_test,predictions4))
print(accuracy_score(y_test,predictions4))
print(classification_report(y_test,predictions4))
```
